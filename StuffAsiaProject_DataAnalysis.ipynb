```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_excel('dataset.xlsx')
dataset = dataset.iloc[:, 0:9]
dataset.to_csv ("dataset.csv", index = None, header=False)
dataset = pd.read_csv('dataset.csv')
dataset.head()
```

```python
# Taking numeric input from columns 
pd.to_numeric(dataset.MPG)
pd.to_numeric(dataset.Cylinders)
pd.to_numeric(dataset.Displacement)
pd.to_numeric(dataset.Horsepower)
pd.to_numeric(dataset.Weight)
pd.to_numeric(dataset.Acceleration)
pd.to_numeric(dataset.Model)
```

```python
# 1) Find the average and total “Acceleration” across the “Origin” column.
average_acc = dataset.groupby(["Origin"]).Acceleration.mean().reset_index()
total_acc = dataset.groupby(["Origin"]).Acceleration.sum().reset_index()
print("average “Acceleration” across the “Origin” column\n\n" ,average_acc)
print("\n\ntotal “Acceleration” across the “Origin” column\n", total_acc)
```

```python
# 3) What is the most frequent and least frequent “Model” for “Origin”?
most_freq_model = dataset.groupby(['Origin']).Model.agg(lambda x:x.value_counts().index[0])
least_freq_model = dataset.groupby(['Origin']).Model.agg(lambda x:x.value_counts().index[-1])
print("most frequent “Model” for “Origin” column\n\n" , most_freq_model)
print("\n\n least frequent “Model” for “Origin” column\n", least_freq_model)
```

```python
# 4) Create a column “isFord”, if the “Car” contains “Ford” or “ford”.
# Assign 1 if the column "isFord" contains “Ford” or “ford”, otherwise 0.
dataset['isFord'] = np.where(dataset.Car.str.contains("Ford|ford"), 1, 0)
dataset.head()
```

```python
# 5) Create a new column (let’s say isNumber). Assign the only digits (ex: 320 for 3rd row) into the column “isNumber”.
import re as re
def find_number(text):
    num = re.findall(r'[0-9]+',text)
    if num:
        return " ".join(num)
    else:
        return "No digits"
dataset['isNumber']=dataset['Car'].apply(lambda x: find_number(x))
dataset.to_csv ("dataset_update.csv", index = None, header=True)
dataset.head()
```

```python
# 6) Calculate the sum and mean of “Weight”
sum_of_weight = dataset.Weight.sum()
mean_of_weight = dataset.Weight.mean()
print("sum of weight: " , sum_of_weight)
print("\nmean of weight: ", mean_of_weight)
```

```python
# 7) Count the total no of “Audi” cars using column “Car”.
number_of_audi = dataset[dataset.Car.str.contains("Audi|audi")]
print("the total no of “Audi” cars\n\n", len(number_of_audi))
```

```python
# 8) Calculate mean and total “Displacement” across “Cylinders”.
total_displacement = dataset.groupby(["Cylinders"]).Displacement.sum().reset_index()
mean_displacement = dataset.groupby(["Cylinders"]).Displacement.mean().reset_index()
print("total “Displacement” across “Cylinders”\n\n" , total_displacement)
print("\n\n mean “Displacement” across “Cylinders”\n", mean_displacement)
```

```python

# 9) Calculate total and mean “Acceleration” by “Model”.
total_acceleration = dataset.groupby(["Model"]).Acceleration.sum().reset_index()
mean_acceleration = dataset.groupby(["Model"]).Acceleration.mean().reset_index()
print("total “Acceleration” by “Model”\n\n" , total_acceleration)
print("\n\n mean “Acceleration” by “Model””\n", mean_acceleration)
```

```python
# 10) Calculate “Origin”-wise mean and total “Horsepower”.
total_housepower = dataset.groupby(["Origin"]).Horsepower.sum().reset_index()
mean_housepower = dataset.groupby(["Origin"]).Horsepower.mean().reset_index()
print("“Origin”-wise total “Horsepower”\n\n" , total_housepower)
print("\n\n “Origin”-wise mean “Horsepower”\n", mean_housepower)
```
